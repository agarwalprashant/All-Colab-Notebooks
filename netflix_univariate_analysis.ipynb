# Netflix Titles Dataset - Detailed Univariate Analysis

Is analysis mein hum Netflix titles dataset ka ek detailed univariate analysis karenge. Har ek variable ko individually examine karenge aur insights nikalenge.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-darkgrid')

# load data
df = pd.read_csv('netflix_titles.csv')
print("Data successfully loaded!")
print(f"Dataset shape: {df.shape}")
```

## 1. Dataset Overview - Basic Information

```python
# basic information of dataset
print("Dataset Columns:")
print(df.columns.tolist())
print("\n" + "="*80 + "\n")

# Check Data types 
print("Data Types:")
print(df.dtypes)
print("\n" + "="*80 + "\n")

# Missing values check
print("Missing Values:")
missing_df = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df) * 100).round(2)
})
print(missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False))
```

```python
# Check First few rows 
print("First 5 rows of the dataset:")
df.head()
```

## 2. Univariate Analysis - Categorical Variables

Ab hum har categorical variable ko detail mein analyze karenge.

### 2.1 Type (Movie vs TV Show)

```python
# Type variable ka analysis
print("Type Distribution:")
type_counts = df['type'].value_counts()
print(type_counts)
print(f"\nPercentage Distribution:")
print(df['type'].value_counts(normalize=True) * 100)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Bar chart
type_counts.plot(kind='bar', ax=ax1, color=['#E50914', '#221f1f'])
ax1.set_title('Netflix Content Type Distribution', fontsize=14, fontweight='bold')
ax1.set_xlabel('Content Type')
ax1.set_ylabel('Count')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0)

# Pie chart
type_counts.plot(kind='pie', ax=ax2, autopct='%1.1f%%', colors=['#E50914', '#221f1f'])
ax2.set_title('Netflix Content Type Percentage', fontsize=14, fontweight='bold')
ax2.set_ylabel('')

plt.tight_layout()
plt.show()
```

### 2.3 Release Year Analysis

```python
# Release year ka analysis
print("Release Year Statistics:")
print(f"Minimum year: {df['release_year'].min()}")
print(f"Maximum year: {df['release_year'].max()}")
print(f"Mean year: {df['release_year'].mean():.2f}")
print(f"Median year: {df['release_year'].median()}")
print(f"Mode year: {df['release_year'].mode()[0] if len(df['release_year'].mode()) > 0 else 'No mode'}")

# Decade-wise distribution
df['decade'] = (df['release_year'] // 10) * 10
decade_counts = df['decade'].value_counts().sort_index()

# Visualizations
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

# Line plot for year-wise trend
yearly_counts = df['release_year'].value_counts().sort_index()
yearly_counts.plot(kind='line', ax=ax1, color='#E50914', linewidth=2)
ax1.set_title('Netflix Content by Release Year', fontsize=16, fontweight='bold')
ax1.set_xlabel('Release Year')
ax1.set_ylabel('Number of Titles')
ax1.grid(True, alpha=0.3)

# Bar plot for decade-wise distribution
decade_counts.plot(kind='bar', ax=ax2, color='#E50914')
ax2.set_title('Netflix Content by Decade', fontsize=16, fontweight='bold')
ax2.set_xlabel('Decade')
ax2.set_ylabel('Number of Titles')
ax2.set_xticklabels([f"{int(d)}s" for d in decade_counts.index], rotation=45)

plt.tight_layout()
plt.show()

# Recent content analysis (last 10 years)
recent_years = df[df['release_year'] >= 2011]['release_year'].value_counts().sort_index()
print(f"\nContent added from last 10 years (2011-2021): {len(df[df['release_year'] >= 2011])} titles")
print(f"Percentage of recent content: {(len(df[df['release_year'] >= 2011]) / len(df) * 100):.2f}%")
```

### 2.4 Country Analysis

```python
# Country analysis - handling multiple countries per entry
print("Country Analysis:")
print(f"Missing country data: {df['country'].isnull().sum()} ({(df['country'].isnull().sum() / len(df) * 100):.2f}%)")

# Pehle primary country extract karte hain (first country in the list)
df['primary_country'] = df['country'].fillna('Unknown').apply(lambda x: x.split(',')[0].strip() if pd.notna(x) and x != 'Unknown' else 'Unknown')

# Top 20 countries
top_countries = df['primary_country'].value_counts().head(20)
print(f"\nTop 20 Content Producing Countries:")
print(top_countries)

# Visualization
plt.figure(figsize=(14, 8))
top_countries.plot(kind='barh', color='#E50914')
plt.title('Top 20 Content Producing Countries on Netflix', fontsize=16, fontweight='bold')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Regional analysis
print("\nRegional Distribution:")
# Simple regional classification
def classify_region(country):
    if country in ['United States', 'Canada', 'Mexico']:
        return 'North America'
    elif country in ['India', 'Japan', 'South Korea', 'China', 'Hong Kong', 'Taiwan', 'Thailand', 'Philippines', 'Indonesia', 'Malaysia', 'Singapore']:
        return 'Asia'
    elif country in ['United Kingdom', 'France', 'Germany', 'Spain', 'Italy', 'Poland', 'Turkey', 'Belgium', 'Netherlands']:
        return 'Europe'
    elif country in ['Brazil', 'Argentina', 'Colombia', 'Chile']:
        return 'South America'
    elif country in ['Australia', 'New Zealand']:
        return 'Oceania'
    elif country in ['Egypt', 'Nigeria', 'South Africa']:
        return 'Africa'
    else:
        return 'Other'

df['region'] = df['primary_country'].apply(classify_region)
region_counts = df['region'].value_counts()
print(region_counts)
```

### 2.5 Duration Analysis

```python
# Duration analysis - Movies aur TV Shows ke liye alag
print("Duration Analysis:")

# Movies ki duration (in minutes)
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration_mins'] = movies_df['duration'].str.extract('(\d+)').astype(float)

print("Movie Duration Statistics (in minutes):")
print(f"Mean duration: {movies_df['duration_mins'].mean():.2f} minutes")
print(f"Median duration: {movies_df['duration_mins'].median()} minutes")
print(f"Min duration: {movies_df['duration_mins'].min()} minutes")
print(f"Max duration: {movies_df['duration_mins'].max()} minutes")
print(f"Std deviation: {movies_df['duration_mins'].std():.2f} minutes")

# TV Shows ki seasons
tv_df = df[df['type'] == 'TV Show'].copy()
tv_df['num_seasons'] = tv_df['duration'].str.extract('(\d+)').astype(float)

print("\nTV Show Season Statistics:")
print(f"Mean seasons: {tv_df['num_seasons'].mean():.2f}")
print(f"Median seasons: {tv_df['num_seasons'].median()}")
print(f"Min seasons: {tv_df['num_seasons'].min()}")
print(f"Max seasons: {tv_df['num_seasons'].max()}")

# Visualizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Movie duration distribution
movies_df['duration_mins'].hist(bins=30, ax=ax1, color='#E50914', edgecolor='black')
ax1.set_title('Distribution of Movie Durations', fontsize=14, fontweight='bold')
ax1.set_xlabel('Duration (minutes)')
ax1.set_ylabel('Frequency')
ax1.axvline(movies_df['duration_mins'].mean(), color='black', linestyle='--', label=f'Mean: {movies_df["duration_mins"].mean():.0f} min')
ax1.legend()

# TV Show seasons distribution
tv_df['num_seasons'].value_counts().sort_index().plot(kind='bar', ax=ax2, color='#221f1f')
ax2.set_title('Distribution of TV Show Seasons', fontsize=14, fontweight='bold')
ax2.set_xlabel('Number of Seasons')
ax2.set_ylabel('Count')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=0)

plt.tight_layout()
plt.show()

# Duration categories for movies
print("\nMovie Duration Categories:")
bins = [0, 60, 90, 120, 150, 300]
labels = ['< 60 min', '60-90 min', '90-120 min', '120-150 min', '> 150 min']
movies_df['duration_category'] = pd.cut(movies_df['duration_mins'], bins=bins, labels=labels)
print(movies_df['duration_category'].value_counts())
```

### 2.6 Genre (Listed In) Analysis

```python
# Genre analysis - multiple genres per entry ko handle karte hain
print("Genre Analysis:")

# Sabhi genres ko extract karte hain
all_genres = []
for genres in df['listed_in'].dropna():
    all_genres.extend([genre.strip() for genre in genres.split(',')])

# Genre frequency
from collections import Counter
genre_counts = Counter(all_genres)
top_genres = pd.Series(dict(genre_counts.most_common(20)))

print(f"Total unique genres: {len(genre_counts)}")
print(f"\nTop 20 Genres:")
print(top_genres)

# Visualization
plt.figure(figsize=(14, 8))
top_genres.plot(kind='barh', color='#E50914')
plt.title('Top 20 Genres on Netflix', fontsize=16, fontweight='bold')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Primary genre analysis (first listed genre)
df['primary_genre'] = df['listed_in'].apply(lambda x: x.split(',')[0].strip() if pd.notna(x) else 'Unknown')
primary_genre_counts = df['primary_genre'].value_counts().head(10)

print("\nTop 10 Primary Genres:")
print(primary_genre_counts)
```

### 2.7 Date Added Analysis

```python
# Date added analysis
print("Date Added Analysis:")
print(f"Missing date_added: {df['date_added'].isnull().sum()} ({(df['date_added'].isnull().sum() / len(df) * 100):.2f}%)")

# Date ko parse karte hain
df['date_added_parsed'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added_parsed'].dt.year
df['month_added'] = df['date_added_parsed'].dt.month
df['day_of_week'] = df['date_added_parsed'].dt.day_name()

# Yearly addition trend
yearly_additions = df['year_added'].value_counts().sort_index()

print(f"\nContent Addition by Year:")
print(yearly_additions.tail(10))  # Last 10 years

# Visualizations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Yearly trend
yearly_additions.plot(kind='bar', ax=ax1, color='#E50914')
ax1.set_title('Netflix Content Additions by Year', fontsize=14, fontweight='bold')
ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Titles Added')

# Monthly pattern
monthly_pattern = df['month_added'].value_counts().sort_index()
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_pattern.index = [month_names[int(i)-1] for i in monthly_pattern.index]
monthly_pattern.plot(kind='bar', ax=ax2, color='#221f1f')
ax2.set_title('Content Addition Pattern by Month', fontsize=14, fontweight='bold')
ax2.set_xlabel('Month')
ax2.set_ylabel('Total Titles Added')

# Day of week pattern
dow_pattern = df['day_of_week'].value_counts()
dow_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
dow_pattern = dow_pattern.reindex(dow_order)
dow_pattern.plot(kind='bar', ax=ax3, color='#E50914')
ax3.set_title('Content Addition Pattern by Day of Week', fontsize=14, fontweight='bold')
ax3.set_xlabel('Day of Week')
ax3.set_ylabel('Total Titles Added')
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=45)

# Growth over time (cumulative)
df_sorted = df.sort_values('date_added_parsed')
df_sorted['cumulative_count'] = range(1, len(df_sorted) + 1)
df_sorted.groupby('year_added')['cumulative_count'].max().plot(kind='line', ax=ax4, color='#E50914', linewidth=3)
ax4.set_title('Cumulative Growth of Netflix Content', fontsize=14, fontweight='bold')
ax4.set_xlabel('Year')
ax4.set_ylabel('Total Titles Available')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Recent additions
print(f"\nContent added in last 3 years: {len(df[df['year_added'] >= 2019])} titles")
print(f"Percentage of recent additions: {(len(df[df['year_added'] >= 2019]) / len(df[df['year_added'].notna()]) * 100):.2f}%")
```

## 3. Text Variables Analysis (Title & Description)

```python
# Title length analysis
df['title_length'] = df['title'].str.len()
df['title_word_count'] = df['title'].str.split().str.len()

print("Title Analysis:")
print(f"Average title length: {df['title_length'].mean():.2f} characters")
print(f"Average word count: {df['title_word_count'].mean():.2f} words")
print(f"Shortest title: '{df.loc[df['title_length'].idxmin(), 'title']}' ({df['title_length'].min()} chars)")
print(f"Longest title: '{df.loc[df['title_length'].idxmax(), 'title']}' ({df['title_length'].max()} chars)")

# Description analysis
df['desc_length'] = df['description'].str.len()
df['desc_word_count'] = df['description'].str.split().str.len()

print("\nDescription Analysis:")
print(f"Average description length: {df['desc_length'].mean():.2f} characters")
print(f"Average word count: {df['desc_word_count'].mean():.2f} words")
print(f"Missing descriptions: {df['description'].isnull().sum()}")

# Visualizations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

# Title length distribution
df['title_length'].hist(bins=30, ax=ax1, color='#E50914', edgecolor='black')
ax1.set_title('Distribution of Title Lengths', fontsize=14, fontweight='bold')
ax1.set_xlabel('Number of Characters')
ax1.set_ylabel('Frequency')

# Title word count distribution
df['title_word_count'].hist(bins=20, ax=ax2, color='#221f1f', edgecolor='black')
ax2.set_title('Distribution of Title Word Count', fontsize=14, fontweight='bold')
ax2.set_xlabel('Number of Words')
ax2.set_ylabel('Frequency')

# Description length distribution
df['desc_length'].hist(bins=30, ax=ax3, color='#E50914', edgecolor='black')
ax3.set_title('Distribution of Description Lengths', fontsize=14, fontweight='bold')
ax3.set_xlabel('Number of Characters')
ax3.set_ylabel('Frequency')

# Description word count distribution
df['desc_word_count'].hist(bins=30, ax=ax4, color='#221f1f', edgecolor='black')
ax4.set_title('Distribution of Description Word Count', fontsize=14, fontweight='bold')
ax4.set_xlabel('Number of Words')
ax4.set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Most common words in titles (simple analysis)
from collections import Counter
import re

# Stop words remove karte hain
stop_words = {'the', 'a', 'an', 'and', 'or', 'but', 'in', 'on', 'at', 'to', 'for', 'of', 'with', 'by', 'is', 'was', 'are', 'were'}

all_words = []
for title in df['title'].dropna():
    words = re.findall(r'\b\w+\b', title.lower())
    all_words.extend([word for word in words if word not in stop_words and len(word) > 2])

word_freq = Counter(all_words)
top_words = pd.Series(dict(word_freq.most_common(20)))

print("\nTop 20 Most Common Words in Titles:")
plt.figure(figsize=(12, 6))
top_words.plot(kind='bar', color='#E50914')
plt.title('Most Common Words in Netflix Titles', fontsize=16, fontweight='bold')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
